Data Cleaning and Preprocessing

```python
# DataFrame is a 2-d datastructure implemented in pandas library
import pandas as pd
```

```python
from sklearn.datasets import fetch_openml
```

```python
# Import dataset using fetch_openml function and store it in a dataframe
df = fetch_openml('titanic', version=1, as_frame=True)['data']
```

```python
df.info()
```

```python
# Is there missing value?  (True  ->  is null)
df.isnull()
```

```python
# How many null value in each column
df.isnull().sum()
```

```python
import seaborn as sns
```

```python
sns.set()
miss_val_per = pd.DataFrame(df.isnull().sum()/len(df)*100)
miss_val_per.plot(kind='bar', title='Missing values in percentage', ylabel='Percentage')
```

```python
df.shape
```

```python
df.drop(columns=['body'],axis=1, inplace=True)
df
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
print(f'Number of null values before imputing: {df.age.isnull().sum()}')
```

```python
# any missing values (NaNs) in the data will be replaced by the mean value of the non-missing values in the same column.
imp = SimpleImputer(strategy='mean')
df['age'] = imp.fit_transform(df[['age']])
print(f'Number of null values before imputing: {df.age.isnull().sum()}')
```

```python
# Impute missing values for whole dataset
def get_parameters(df):
    parameters = {}
    for col in df.columns[df.isnull().any()]:
        if df[col].dtype in ['float64', 'int64', 'int32']:
            strategy = 'mean'
        else:
            strategy = 'most_frequent'
        parameters[col] = {'strategy': strategy}
    return parameters
get_parameters(df)
```

```python
parameters = get_parameters(df)
for col, param in parameters.items():
    strategy = param['strategy']
    imp = SimpleImputer(strategy=strategy)
    df[col] = imp.fit_transform(df[[col]]).ravel()
```
